```python
def gen_data(date):
    url='https://music.bugs.co.kr/chart/track/day/total?chartdate='+str(date)
    html=requests.get(url)
    soup=bs(html.text, 'html.parser')
    tbody=soup.find("tbody")
    trsoup=tbody.find_all("tr")
    data=[]
    for song in trsoup:
        date=date
        ranking=song.find('div', class_='ranking').text.split('\n')[1]
        title=song.find('p', class_='title').text.replace("\n","")
        artist=song.find('p', class_='artist').text.replace("\n","")
        album=song.find('a', class_='album').text.replace("\n","")
        data.append({"날짜" : date, "랭킹": ranking, "곡명" : title, "아티스트" : artist, "앨범명" : album})
    data=pd.DataFrame(data)
    return data

def check_date(date):
    if int(date) > int( datetime.datetime.strftime(datetime.datetime.today(),'%Y%m%d') ):
        date=int( datetime.datetime.strftime(datetime.datetime.today(),'%Y%m%d') )
        print(f"검색 가능한 끝 날짜는 어제까지입니다. 대체된 날짜는 {date}입니다")
    return date
    
df=pd.DataFrame()
sdate=int(input("시작 날짜를 입력해주세요\n"))
edate=int(input("끝 날짜를 입력해주세요\n"))

edate=check_date(edate)
tmp_date=sdate

while(int(tmp_date)<=int(edate)):
    data=gen_data(tmp_date)
    df=pd.concat([df,data])
    tmp_date=datetime.datetime.strptime(str(tmp_date), '%Y%m%d')+relativedelta(days=1)
    tmp_date=int( datetime.datetime.strftime(tmp_date,'%Y%m%d') )

df.reset_index(drop=True, inplace=True)
df
```
